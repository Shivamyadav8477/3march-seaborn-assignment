```python
Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
```

```python
Seaborn is a Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating informative and attractive statistical graphics. Seaborn offers various types of plots, each with its specific use cases. Here are five commonly used plots in Seaborn and their uses:

1. **Histograms (`sns.histplot()`):**
   - **Use**: Histograms are used to visualize the distribution of a single variable's values. They help in understanding the data's frequency distribution, central tendency, and spread.
   
2. **Scatter Plots (`sns.scatterplot()`):**
   - **Use**: Scatter plots are used to visualize the relationship between two continuous variables. They help in identifying patterns, trends, correlations, and outliers in the data.

3. **Bar Plots (`sns.barplot()` or `sns.countplot()`):**
   - **Use**: Bar plots are used to compare categorical data or visualize the distribution of a categorical variable. `sns.barplot()` is used for showing the mean or aggregate values, while `sns.countplot()` shows the count of occurrences of each category.

4. **Box Plots (`sns.boxplot()`):**
   - **Use**: Box plots are used to display the distribution and spread of data, including the median, quartiles, and potential outliers. They are useful for identifying data variability and comparing distributions across categories.

5. **Heatmaps (`sns.heatmap()`):**
   - **Use**: Heatmaps are used to visualize the relationships between two categorical variables. They display the data in a grid format, with colors representing the strength or value of the relationship. Heatmaps are often used in correlation matrices or to show patterns in matrices of data.

These are just a few examples of the plots available in Seaborn. Seaborn also provides tools for creating violin plots, pair plots, regression plots, and more. The choice of plot depends on the specific data and the insights you want to gain from it.
```

```python
Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
```

```python
To load the "fmri" dataset using Seaborn's load_dataset function and create a line plot for different events and regions, you can follow these steps
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Create a line plot using Seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(data=fmri_data, x="timepoint", y="signal", hue="event", style="region")

# Set plot title and labels
plt.title("Line Plot of FMRI Data by Event and Region")
plt.xlabel("Timepoint")
plt.ylabel("Signal")

# Show the legend
plt.legend(title="Event")

# Display the plot
plt.show()
```

```python
In this code:

We load the "fmri" dataset using sns.load_dataset("fmri") and store it in the fmri_data variable.

We create a line plot using sns.lineplot() and specify the x, y, hue, and style parameters. We use "timepoint" for the x-axis, "signal" for the y-axis, "event" for coloring different events, and "region" for differentiating regions with different line styles.

We set the plot title, x-label, and y-label using plt.title(), plt.xlabel(), and plt.ylabel().

We show the legend to differentiate between different events using plt.legend().

Finally, we use plt.show() to display the line plot.

This code will create a line plot showing how the "signal" varies with "timepoint" for different events and regions in the "fmri" dataset. The legend will help you identify different events.
```

```python
Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.
```

```python
To load the "titanic" dataset using Seaborn's load_dataset function and create two box plots, one for 'age' and one for 'fare', with 'pclass' on the x-axis, you can use the following code:
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")

# Create subplots for two box plots side by side
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Box plot for 'age' vs 'pclass'
sns.boxplot(data=titanic_data, x='pclass', y='age', ax=axes[0])
axes[0].set_title('Box Plot of Age by Pclass')
axes[0].set_xlabel('Pclass')
axes[0].set_ylabel('Age')

# Box plot for 'fare' vs 'pclass'
sns.boxplot(data=titanic_data, x='pclass', y='fare', ax=axes[1])
axes[1].set_title('Box Plot of Fare by Pclass')
axes[1].set_xlabel('Pclass')
axes[1].set_ylabel('Fare')

# Adjust the layout
plt.tight_layout()

# Show the plots
plt.show()
```

```python
In this code:

We load the "titanic" dataset using sns.load_dataset("titanic") and store it in the titanic_data variable.

We create subplots with two columns (one for 'age' and one for 'fare') using plt.subplots(1, 2).

For the first box plot ('age' vs 'pclass'), we use sns.boxplot() with 'pclass' on the x-axis and 'age' on the y-axis. We set the title, x-label, and y-label for this subplot.

For the second box plot ('fare' vs 'pclass'), we use sns.boxplot() with 'pclass' on the x-axis and 'fare' on the y-axis. We set the title, x-label, and y-label for this subplot as well.

We adjust the layout using plt.tight_layout() to prevent overlapping.

Finally, we use plt.show() to display the two box plots side by side.

This code will create two box plots that compare 'age' and 'fare' by 'pclass' in the "titanic" dataset. The first box plot shows the distribution of ages across different passenger classes, and the second one shows the distribution of fares across different passenger classes.
```

```python
Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.
```

```python
To plot a histogram for the 'price' column of the "diamonds" dataset from Seaborn and use the hue parameter to differentiate by the 'cut' column, you can use the following code:
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds_data = sns.load_dataset("diamonds")

# Create a histogram with hue for the 'cut' column
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_data, x='price', hue='cut', bins=30, kde=True)
plt.title('Histogram of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Frequency')

# Show the legend
plt.legend(title='Cut')

# Display the plot
plt.show()
```

```python
In this code:

We load the "diamonds" dataset using sns.load_dataset("diamonds") and store it in the diamonds_data variable.

We create a histogram using sns.histplot() with 'price' on the x-axis and 'cut' as the hue parameter to differentiate by the 'cut' column. We specify the number of bins as 30 and include a kernel density estimate (KDE) curve for better visualization.

We set the plot title, x-label, and y-label using plt.title(), plt.xlabel(), and plt.ylabel().

We display the legend to differentiate between different cuts using plt.legend().

Finally, we use plt.show() to display the histogram.

This code will create a histogram of diamond prices with different colors representing different 'cut' categories in the "diamonds" dataset. The hue parameter allows you to see how the price distribution varies among different cuts.
```

```python
To plot a pair plot for the "iris" dataset from Seaborn and use the hue parameter to differentiate by the "species" column, you can use the following code:
```

```python
Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights_data = sns.load_dataset("flights")

# Pivot the dataset to create a matrix for the heatmap
flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(flights_pivot, annot=True, fmt='d', cmap='coolwarm', linewidths=1, linecolor='black')
plt.title('Heatmap of Flights Data')
plt.xlabel('Year')
plt.ylabel('Month')

# Display the plot
plt.show()
```

```python
In this code:

We load the "flights" dataset using sns.load_dataset("flights") and store it in the flights_data variable.

To create a heatmap, we need to pivot the dataset using pivot_table() to create a matrix where rows represent months, columns represent years, and values represent the number of passengers.

We create the heatmap using sns.heatmap() with the annot parameter set to True to display the values in the cells, fmt set to 'd' for integer formatting, cmap set to 'coolwarm' for the color map, linewidths set to 1 to add lines between cells, and linecolor set to 'black' for the line color.

We set the plot title, x-label, and y-label using plt.title(), plt.xlabel(), and plt.ylabel().

Finally, we use plt.show() to display the heatmap.

This code will create a heatmap of the "flights" dataset, showing the number of passengers for different months and years. The color of each cell represents the passenger count, and annotations display the exact values in each cell.
```
